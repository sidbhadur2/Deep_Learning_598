```python
# Library Import
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import csv
```

```python
# Get current working directory
cwd = os.getcwd()
print(cwd)
```

```python
input1 ='Learning_Curve_Seaquest_1.csv'
input2 ='Learning_Curve_Seaquest_2.csv'
input3 ='Learning_Curve_Seaquest_3.csv'
beta_zero = pd.read_csv(input1,header=None)
beta_point_zero_one = pd.read_csv(input2,header=None)
beta_one = pd.read_csv(input3,header=None)
```

```python
beta_zero[1] = beta_zero[1].str.strip(']').astype(int)
```

```python
beta_point_zero_one[1] = beta_point_zero_one[1].str.strip(']').astype(int)
```

```python
beta_one[1] = beta_one[1].str.strip(']').astype(int)
```

```python
beta_zero_df = beta_zero[[1,7]]
beta_zero_df.columns = ['Time', 'RScore']
#beta_zero_df['Time'] = beta_zero_df['Time'].str.strip(']').astype(int)
#beta_zero_df
```

```python
beta_zero_one_df = beta_point_zero_one[[1,7]]
beta_zero_one_df.columns = ['Time', 'RScore']
#beta_zero_one_df
```

```python
beta_one_df = beta_one[[1,7]]
beta_one_df.columns = ['Time', 'RScore']
#beta_one_df
```

```python
x1 = beta_zero_df['Time']
y1 = beta_zero_df['RScore']
plt.plot(x1, y1)
plt.ylabel('RScore')
plt.xlabel('Time')
plt.title('RL Seaquest Game Beta = 0')
plt.savefig('RL_1.png')
plt.show()
```

```python
x2 = beta_zero_one_df['Time']
y2 = beta_zero_one_df['RScore']
plt.plot(x2, y2)
plt.ylabel('RScore')
plt.xlabel('Time')
plt.title('RL Seaquest Game Beta = 0.01')
plt.savefig('RL_2.png')
plt.show()
```

```python
x3 = beta_one_df['Time']
y3 = beta_one_df['RScore']
plt.plot(x3, y3)
plt.ylabel('RScore')
plt.xlabel('Time')
plt.title('RL Seaquest Game Beta = 1')
plt.savefig('RL_3.png')
plt.show()
```
